```python
#x = df[['symmetry error', 'texture error', 'mean fractal dimension', 'smoothness error']].values
x = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
xt,xs,yt,ys = tts(x,y,test_size=0.25,random_state=101)
```

```python
s = MinMaxScaler()
```

```python
s.fit(xt)
```

```python
xt = s.transform(xt)
```

```python
xs = s.transform(xs)
```

```python
xs.max()
```

```python
m = Sequential()

m.add(Dense(60,activation='relu'))
m.add(Dense(30,activation='relu'))
m.add(Dense(60,activation='relu'))
m.add(Dense(15,activation='relu'))
m.add(Dense(1,activation='sigmoid'))

m.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
m.fit(xt,yt,epochs=500,validation_data=(xs,ys))
```

```python
r = pd.DataFrame(m.history.history)
r.plot()
```

```python
p = m.predict(xs)
m.evaluate(xs,ys,verbose=0)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
m = Sequential()

m.add(Dense(60,activation='relu'))
m.add(Dense(30,activation='relu'))
m.add(Dense(60,activation='relu'))
m.add(Dense(15,activation='relu'))
m.add(Dense(1,activation='sigmoid'))

m.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
es = EarlyStopping(monitor='val_loss', mode="min", verbose=1, patience=25)
```

```python
m.fit(xt,yt,epochs=600,validation_data=(xs,ys),callbacks=[es])
```

```python
r = pd.DataFrame(m.history.history)
r.plot()
```

```python
p = m.predict(xs)
m.evaluate(xs,ys,verbose=0)
```

```python
mean_squared_error(ys,p)**0.5
```

```python
mean_absolute_error(ys,p)
```

```python
explained_variance_score(ys,p)
```

```python
from tensorflow.keras.layers import Dropout
```

```python
m = Sequential()

m.add(Dense(60,activation='relu'))
m.add(Dropout(0.5))
m.add(Dense(30,activation='relu'))
m.add(Dense(60,activation='relu'))
m.add(Dense(15,activation='relu'))
m.add(Dense(1,activation='sigmoid'))

m.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
m.fit(xt,yt,epochs=600,validation_data=(xs,ys),callbacks=[es])
```

```python
r = pd.DataFrame(m.history.history)
r.plot()
```

```python
p = m.predict(xs)
p = (p >= 0.5).astype(int)
m.evaluate(xs,ys,verbose=0)
```

```python
mean_squared_error(ys,p)**0.5
```

```python
mean_absolute_error(ys,p)
```

```python
explained_variance_score(ys,p)
```
