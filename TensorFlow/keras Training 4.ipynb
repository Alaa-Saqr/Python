```python
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard, ModelCheckpoint

# Create directories
log_dir = "logs/fit/"
model_dir = "models/"
os.makedirs(log_dir, exist_ok=True)
os.makedirs(model_dir, exist_ok=True)

# TensorBoard callback
b = TensorBoard(log_dir=ld,
                histogram_freq=1,
                write_graph=True,
                write_images=True,
                update_freq='epoch',
                profile_batch=2,
                embeddings_freq=1)

# Model checkpoint callback
checkpoint = ModelCheckpoint(filepath=model_dir + "best_model.keras",
                             monitor='val_loss',
                             save_best_only=True,
                             verbose=1)

# Define the model
m = Sequential()
m.add(Dense(60, activation='relu'))
m.add(Dense(30, activation='relu'))
m.add(Dropout(0.5))
m.add(Dense(60, activation='relu'))
m.add(Dense(30, activation='relu'))
m.add(Dense(15, activation='relu'))
m.add(Dense(1, activation='sigmoid'))

# Compile the model
m.compile(optimizer='adam', loss='binary_crossentropy')

# Early stopping
es = EarlyStopping(monitor='val_loss', mode="min", verbose=1, patience=25)

# Fit the model
m.fit(xt, yt, epochs=600, validation_data=(xs, ys), callbacks=[es, b, checkpoint])

```
