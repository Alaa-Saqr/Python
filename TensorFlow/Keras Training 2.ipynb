```python
x = df.drop('price',axis=1).values
y = df['price'].values
```

```python
%run "C:/s/ker.py"
```

```python
s = MinMaxScaler()
```

```python
xt,xs,yt,ys = tts(x,y,test_size=0.3,random_state=101)
```

```python
#xt,xv,yt,yv = tts(x,y,test_size=0.3,random_state=101)
```

```python
s.fit(xt)
```

```python
xt = s.transform(xt)
```

```python
xs = s.transform(xs)
```

```python
xt.shape
```

```python
m = Sequential()

m.add(Dense(36,activation='relu'))
m.add(Dense(18,activation='relu'))
m.add(Dense(18,activation='relu'))
m.add(Dense(18,activation='relu'))
m.add(Dense(1,))

m.compile(optimizer='adam',loss='mse')
```

```python
m.fit(xt,yt,validation_data=(xs,ys),epochs=5)
```

```python
r = pd.DataFrame(m.history.history)
r.plot()
```

```python
m.evaluate(xs,ys,verbose=0)
```

```python
p = m.predict(xs)
```

```python
mean_squared_error(ys,p)**0.5
```

```python
mean_absolute_error(ys,p)
```
